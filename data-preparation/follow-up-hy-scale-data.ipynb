```python
import pandas as pd
import numpy as np
```

```python
stageFile = "../data/HY_Baseline_Stage.csv"
volumes = "../data/volumes.csv"

stageDf = pd.read_csv(stageFile)
volumesDf = pd.read_csv(volumes)
stageDf = stageDf[["PATNO", "EVENT_ID","hy_on"]]
initialStage = []
followUpStage = []

for subId in volumesDf["subjectId"].values:
    if (len(stageDf.loc[(stageDf['PATNO'] == subId) & (stageDf['EVENT_ID'] == "BL")]["hy_on"]) > 0):
        initialStage.append(stageDf.loc[(stageDf['PATNO'] == subId) & (stageDf['EVENT_ID'] == "BL")]["hy_on"].values[0])
    else:
        initialStage.append(-1)
        
    if (len(stageDf.loc[(stageDf['PATNO'] == subId) & (stageDf['EVENT_ID'] == "V08")]["hy_on"]) > 0):
        followUpStage.append(stageDf.loc[(stageDf['PATNO'] == subId) & (stageDf['EVENT_ID'] == "V08")]["hy_on"].values[0])
    else:
        followUpStage.append(-1)
    
volumesDf["initialHY"] = initialStage
volumesDf["followUpHY"] = followUpStage
volumesDf = volumesDf[volumesDf.followUpHY != -1]
volumesDf = volumesDf[volumesDf.initialHY != -1]
volumesDf = volumesDf.dropna(subset=['initialHY', 'followUpHY'])
volumesDf = volumesDf.drop(["stage"], 1)
volumesDf
```
